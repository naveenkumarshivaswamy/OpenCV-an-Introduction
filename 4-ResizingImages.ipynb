```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
#reading the image 
image = cv2.imread('resizing.jpg', cv2.IMREAD_UNCHANGED) 
plt.imshow(image)
```

```python
print('Original Dimensions : ',image.shape)
```

```python
scale_percent = 60 # percent of original size
width = int(image.shape[1] * scale_percent / 100)
height = int(image.shape[0] * scale_percent / 100)
dim = (width, height)
print(dim)
```

```python
# NTER_NEAREST: Nearest neighbor interpolation
# INTER_LINEAR: Bilinear interpolation
# INTER_AREA: Resampling using pixel area relation
# INTER_CUBIC: Bicubic interpolation over 44 pixel neighborhood
# INTER_LANCZOS4: Lanczos interpolation over 88 neighborhood
```

```python
# #converting image to size (100,100,3) 
# smaller_image = cv2.resize(image,dim, interpolation = cv2.NTER_NEAREST)
# #plot the resized image
# plt.imshow(smaller_image)
```

```python
#converting image to size (100,100,3) 
smaller_image = cv2.resize(image,dim, interpolation = cv2.INTER_LINEAR)
#plot the resized image
plt.imshow(smaller_image)
```

```python
#converting image to size (100,100,3) 
smaller_image = cv2.resize(image,dim, interpolation = cv2.INTER_AREA)
#plot the resized image
plt.imshow(smaller_image)
```

```python
#converting image to size (100,100,3) 
smaller_image = cv2.resize(image,dim, interpolation = cv2.INTER_CUBIC)
#plot the resized image
plt.imshow(smaller_image)
```

```python
#converting image to size (100,100,3) 
smaller_image = cv2.resize(image,dim, interpolation = cv2.INTER_LANCZOS4)
#plot the resized image
plt.imshow(smaller_image)
```
