```python
#import required libraries
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 
%matplotlib inline 
```

```python
#read the image
image = cv2.imread('coins.jpg')

#converting image to grayscale format
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

#apply thresholdin 
ret,mask = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU) 
plt.imshow(mask)
```

```python
#apply AND operation on image and mask generated by thrresholding
final = cv2.bitwise_and(image,image,mask = mask) 
#plot the result
plt.imshow(final)
```

```python
#apply NOT operation on image and mask generated by thrresholding
final = cv2.bitwise_not(image,image,mask = mask) 
#plot the result
plt.imshow(final)
```

```python
#apply OR operation on image and mask generated by thrresholding
final = cv2.bitwise_or(image,image,mask = mask) 
#plot the result
plt.imshow(final)
```

```python
#apply XOR operation on image and mask generated by thrresholding
final = cv2.bitwise_xor(image,image,mask = mask) 
#plot the result
plt.imshow(final)
```
